###배열의 축을 확인하는 코드 shape




```python
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([[1, 2, 3, 4, 5]])
z = np.array([[1], [2], [3], [4], [5]])

print(x.shape, y.shape, z.shape)
```

###배열 생성 함수

####arange() 함수: 주어진 범위내의 연속된 값은 가지는 배열 생성


```python
x = np.arange(0, 10)
y = np.arange(0, 10, 0.5)
z = np.arange(10, 0, -1)
a = np.arange(0 ,34, 3)
print(x)
print(y)
print(z)
print(a)
```

####zeros() 함수: 주어진 형태의 배열을 만들고 0으로 채움


```python
x = np.zeros(5)
y = np.zeros((2, 4))
print(x, x.shape)
print(y, y.shape)
```

####full() 함수: 주어진 형태의 shape만큼 특정값으로 채움


```python
x = np.full((2, 3), 0.5)
print(x)
```

#####ones_like(), zeros_like() 함수


```python
x = np.array([[1, 2], [3, 4]])
print(x.shape)

y = np.zeros_like(x)
print(y, y.shape)

z = np.ones_like(x)
print(z, z.shape)
```

###스타일 지정하기
**fmt = "[color][marker][line]"**

*   선 색상: 'b', 'r', ...
*   마커 모양: '.', 'o', '^', '*'
*   선 스타일: '-', '--', '-.', '.'




```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x+0.5)
y3 = np.sin(x+1)

plt.plot(x, y1, 'b^:')
plt.plot(x, y2, 'ko-')
plt.plot(x, y3, 'r*-')
plt.show()
```

```python
x = np.arange(10) #값이 하나만 있으면 목적지값
plt.plot(x, x, 'r.--')
plt.plot(x, x+2, 'bo-')
plt.plot(x, x+4, 'gv-.')
plt.show()
```

```python
x = np.arange(10) #값이 하나만 있으면 목적지값
plt.plot(x, x, 'r.--', markersize = 40)
plt.plot(x, x+2, 'bo-', linewidth = 3) #라인의 두께
plt.plot(x, x+4, 'gv-.', linewidth = 3, markersize = 5)
plt.show()
```

###그래프 형상 설정

```python
#범례 위치 설정
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x + 0.5)
y3 = np.sin(x + 1)

plt.plot(x, y1, 'r--', label='Sin A')
plt.plot(x, y2, 'b.', label='Sin B')
plt.plot(x, y3, 'g-.', label='Sin C')
plt.legend(loc='upper right')
plt.show()
```

```python
#눈금 설정 ticks()
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, 'b-.', linewidth=3)
plt.xticks(np.arange(0, 10, 2))
plt.yticks(np.arange(-1.0, 1,1))
plt.yticks(np.arange(-1.0, 1.1))
plt.show()
```

###그래프의 구조

```python
#그래프 크기 설정: figure() (inch 단위)
x = np.arange(1, 30)
y = np.log(x)

plt.figure(figsize = (4, 3))
plt.plot(x, y)
plt.show()
```

###여러 개의 그래프 그리기

```python
#subplot(): 여러 개의 그래프 그리기
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(x, y, 'r-')
plt.subplot(2, 2, 2)
plt.plot(x, z, 'g--')
plt.subplot(2, 2, 3)
plt.plot(x, z, 'b-.')
plt.show()
```

###산점도(Scatter Plot)

```python
x = np.random.rand(30)
y = np.random.rand(30)

plt.scatter(x, y)
plt.show()
```

```python
n = 50
x = np.random.rand(n)
y = np.random.rand(n)
area = (30 * np.random.rand(n))**2
colors = np.random.rand(n)

#alpha는 겸치는 부분의 투명
plt.scatter(x, y, s=area, c=colors, alpha=0.7)
plt.show()
```

###그래프 파일 저장
